```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 设置Chrome选项
chrome_options = Options()
chrome_options.add_argument("--headless=new")  # 新的无头模式
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# ChromeDriver路径
chrome_driver_path = r"C:\Users\Administrator\Desktop\X\chrome-win64\chromedriver.exe"  # 使用原始字符串

# 启动ChromeDriver
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Twitter用户的URL
twitter_url = 'https://x.com/elonmusk'

# 存储最新的推文ID
last_tweet_id = None

def get_latest_tweet_info():
    global last_tweet_id
    driver.get(twitter_url)
    
    # 使用WebDriverWait确保页面加载完成
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, 'article'))
    )

    # 查找所有推文文章元素
    tweets = driver.find_elements(By.CSS_SELECTOR, 'article')
    if not tweets:
        print("No tweets found on the page.")
        return None, None

    latest_tweet = tweets[0]
    
    # 查找推文URL
    tweet_link_elements = latest_tweet.find_elements(By.CSS_SELECTOR, 'a')
    tweet_url = None
    for link in tweet_link_elements:
        href = link.get_attribute('href')
        if 'status' in href:
            tweet_url = href
            break

    # 查找推文内容
    tweet_content_element = latest_tweet.find_element(By.CSS_SELECTOR, 'div[lang]')
    tweet_content = tweet_content_element.text if tweet_content_element else "No content found"
    
    # 检查是否为新推文
    if tweet_url:
        tweet_id = tweet_url.split('/')[-1]
        if tweet_id != last_tweet_id:
            last_tweet_id = tweet_id
            return tweet_url, tweet_content
    
    return None, None

# 获取最新推文URL和内容
latest_tweet_url, latest_tweet_content = get_latest_tweet_info()
if latest_tweet_url:
    print(f"New Tweet URL: {latest_tweet_url}")
    print(f"New Tweet Content: {latest_tweet_content}")
else:
    print("No new tweet found.")

# 关闭WebDriver
driver.quit()
```
